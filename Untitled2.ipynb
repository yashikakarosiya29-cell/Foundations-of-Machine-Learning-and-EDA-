Question 2: Explain overfitting and underfitting in ML. How can you detect and prevent them ?
Hint: Discuss bias-variance tradeoff, cross-validation, and regularization techniques.

- **Overfitting and Underfitting in Machine Learning**
1. **Overfitting**

. Definition: Overfitting occurs when a machine learning model learns the training data too well, including its noise and outliers, resulting in poor performance on new, unseen data.

. Detection:

- High accuracy on training data but low accuracy on test/validation data.

- Large gap between training and validation error.

. Prevention Techniques:

- Cross-validation: Use k-fold cross-validation to ensure model generalizes well.

- Regularization: Apply L1 (Lasso) or L2 (Ridge) regularization to penalize large weights.

- Simplify the model: Reduce model complexity (fewer features or layers).

- More data: Increase training data to help the model generalize better.

**2. Underfitting**

. Definition: Underfitting occurs when a model is too simple to capture the underlying patterns in the data, resulting in poor performance on both training and test data.

. Detection:

- Low accuracy on both training and validation/test datasets.

- High bias in the model.

. Prevention Techniques:

- Increase model complexity: Use more features, deeper models, or advanced algorithms.

- Feature engineering: Add relevant features or transform existing ones.

- Reduce regularization: If regularization is too strong, the model may underfit.

**3. Bias-Variance Tradeoff**

- Bias: Error due to overly simple models → leads to underfitting.

- Variance: Error due to overly complex models → leads to overfitting.

- Goal: Find a balance between bias and variance for optimal generalization.






Question 3:How would you handle missing values in a dataset? Explain at least three methods with examples.
Hint: Consider deletion, mean/median imputation, and predictive modeling.
- **Handling Missing Values in a Dataset**

Missing values are common in real-world datasets and need proper handling, as they can affect model accuracy. Here are three main methods:
1. Deletion (Removing Missing Data)

Method: Remove rows or columns with missing values.
Example:

- If a dataset has 1000 rows and 10 rows have missing age values, we can remove those 10 rows.

Python:

df.dropna(inplace=True)  # drops rows with missing values


Pros: Simple to implement.
Cons: Can lead to loss of valuable data if many values are missing.

2. Mean/Median/Mode Imputation

Method: Replace missing values with the mean, median, or mode of the column.
Example:

- If the “Age” column has missing values, replace them with the mean age:

 df['Age'].fillna(df['Age'].mean(), inplace=True)


  Median is preferred for skewed data. Mode is used for categorical data.
. Pros: Easy to implement, keeps dataset size intact.
. Cons: Can reduce variance and may introduce bias.

3. Predictive Modeling (Imputation using ML)

Method: Predict missing values using other features in the dataset.
Example:

Use regression to predict missing “Salary” based on “Experience” and “Education”:

from sklearn.linear_model import LinearRegression

 . train model on rows with known Salary
model = LinearRegression()
model.fit(X_train, y_train)

 . predict missing Salary
df.loc[df['Salary'].isna(), 'Salary'] = model.predict(X_missing)


Pros: More accurate than mean/median imputation.
Cons: More complex and time-consuming.


Question 4:What is an imbalanced dataset? Describe two techniques to handle it (theoretical + practical).
Hint: Discuss SMOTE, Random Under/Oversampling, and class weights in models.
- **Imbalanced Dataset and Handling Techniques**
An imbalanced dataset occurs when the classes in a classification problem are not represented equally.

Example: In a fraud detection dataset, 95% of transactions may be legitimate and only 5% fraudulent.

Problem: Models may bias toward the majority class, resulting in poor performance for the minority class.

**Techniques to handle it:**

- Resampling:

Oversampling: Duplicate or generate minority class samples (e.g., SMOTE).

Undersampling: Remove majority class samples to balance the dataset.

- Class Weights:

Assign higher weight to the minority class in models (e.g., class_weight='balanced' in Logistic Regression).

These techniques help the model perform better on the minority class without being biased.




Question 5: Why is feature scaling important in ML? Compare Min-Max scaling and standerization ?

Hint: Explain impact on distance-based algorithms (e.g., KNN, SVM) and gradient

- **Feature Scaling in Machine Learning**
1. Why is Feature Scaling Important?

Feature scaling is the process of normalizing or standardizing the range of features so that all features contribute equally to the model.

Many ML algorithms, especially distance-based (KNN, SVM, K-Means) and gradient-based (Gradient Descent in Linear/Logistic Regression, Neural Networks), are sensitive to feature scale.

Without scaling, features with larger values can dominate the learning process, leading to biased results and slower convergence.

2. Common Techniques
a) Min-Max Scaling (Normalization)

Rescales features to a fixed range [0,1].

Pros: Preserves original distribution, easy interpretation.

Cons: Sensitive to outliers.

Example in Python:
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

b) Standardization (Z-score Scaling)

Rescales features to have mean = 0 and standard deviation = 1.

Example in Python:

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)








Question 6: Compare Label Encoding and One-Hot Encoding. When would you prefer one over the other?
Hint: Consider categorical variables with ordinal vs. nominal relationships.

-**Label Encoding vs One-Hot Encoding**
1. Label Encoding

- Converts each category into a unique integer value.

- Example: ["Red", "Green", "Blue"] → [0, 1, 2]

- Pros: Simple and memory-efficient.

- Cons: Introduces ordinal relationship even if categories are nominal, which can mislead some algorithms.

- Use Case: Best for ordinal categorical variables (e.g., Low < Medium < High).

2. One-Hot Encoding

- Converts each category into a binary vector (0 or 1).

- Example: ["Red", "Green", "Blue"] → [[1,0,0],[0,1,0],[0,0,1]]

- Pros: Avoids ordinal relationship, suitable for nominal categories.

- Cons: Can increase dimensionality if many categories exist.

- Use Case: Best for nominal categorical variables (e.g., Color, Country).


Question 7: Google Play Store Dataset
a). Analyze the relationship between app categories and ratings. Which categories have the highest/lowest average ratings, and what could be the possible reasons?
Dataset: https://github.com/MasteriNeuron/datasets.git
(Include your Python code and output in the code box below.)



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
url = "https://raw.githubusercontent.com/MasteriNeuron/datasets/main/googleplaystore.csv"
df = pd.read_csv(url)

# Check first few rows
df.head()

```

```python
df = df.dropna(subset=['Category', 'Rating'])
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')
df = df.dropna(subset=['Rating'])

```

```python
df = df[df['Rating'] <= 5]
category_ratings = df.groupby('Category')['Rating'].mean().sort_values(ascending=False)
category_ratings

```

```python
highest_category = category_ratings.idxmax()
highest_rating = category_ratings.max()

lowest_category = category_ratings.idxmin()
lowest_rating = category_ratings.min()

print(f"Highest rated category: {highest_category} ({highest_rating:.2f})")
print(f"Lowest rated category: {lowest_category} ({lowest_rating:.2f})")
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=category_ratings.index, y=category_ratings.values, palette="viridis")
plt.xticks(rotation=90)
plt.ylabel('Average Rating')
plt.title('Average App Ratings by Category')
plt.show()
```

**Analyze and Explain**

- Highest rated categories: Likely niche/specialized apps (Events, Education, Art & Design)

Users have clear expectations → higher satisfaction

Fewer apps → ratings skew positive

-  Lowest rated categories: Likely broad/popular apps (Communication, Social, Tools)

Large user base → mixed feedback

App crashes, ads, bugs → lower ratings

High competition → quality varies
